```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Load breast cancer dataset
cancer_data = datasets.load_breast_cancer()

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, test_size=0.3, random_state=109)

# Initialize Perceptron model
perceptron = Perceptron()

# Train the model
perceptron.fit(X_train, y_train)

# Predict labels for test set
y_pred = perceptron.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calculate recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Gim chiu d liu xung cn 2 chiu
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Hin th kt qu d liu gim chiu
plt.figure(figsize=(8, 6))
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap=plt.cm.coolwarm, s=50, edgecolors='k')
plt.title('Breast Cancer Data after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Class')
plt.grid(True)
plt.show()

# Phn loi bng Logistic Regression
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train_pca, y_train)
y_pred_logistic = logistic_regression.predict(X_test_pca)

# nh gi kt qu vi Logistic Regression
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
precision_logistic = precision_score(y_test, y_pred_logistic)
recall_logistic = recall_score(y_test, y_pred_logistic)

print("Logistic Regression:")
print("Accuracy:", accuracy_logistic)
print("Precision:", precision_logistic)
print("Recall:", recall_logistic)

# So snh vi kt qu ca Perceptron
print("\nPerceptron:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```
