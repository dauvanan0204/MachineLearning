```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
 
# Change to data path on your computer 
data = pd.read_csv("D:\\Python\\MachineLearning\\MachineLearning\\week02\\data\\SAT_GPA.csv") 
# Show the description of data 
data.describe() 
 
# Set to training data (x, y) 
y = data['GPA'] 
x = data['SAT'] 
# Remind that we need to put component x_0 = 1 to x 
plt.scatter(x,y) 
plt.xlabel('SAT', fontsize = 20) 
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x_train = np.array([x[:60]]).T
y_train = np.array([y[:60]]).T
x_val = np.array([x[60:]]).T
y_val = np.array([y[60:]]).T

one = np.ones((x_train.shape[0], 1))
Xbar = np.concatenate((one, x_train), axis = 1)
# Calculating weights of the fitting line
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y_train)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)

# Preparing the fitting line
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(1634, 2050, 2)
# sat_min = data['SAT'].min()
# sat_max = data['SAT'].max()
# print(sat_max)
# print(sat_min)
y0 = w_0 + w_1*x0

# Drawing the fitting line
plt.scatter(x_train,y_train)
fig = plt.plot(x0,y0, lw=4, c='orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x_val = np.array([x[60:]]).T
y_val = np.array([y[60:]]).T
# Tính toán dữ liệu dự đoán trên dữ liệu validation
x_val_ones = np.concatenate((np.ones((x_val.shape[0], 1)), x_val), axis=1)
y_val_pred = np.dot(x_val_ones, w)
# print(y_val_pred)
# Tính toán tổng bình phương sai số trên dữ liệu validation
mse_val = np.mean((y_val - y_val_pred)**2)
print("Tổng bình phương sai số trên dữ liệu validation:", mse_val)
```
